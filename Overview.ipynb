```python
def iterate(n):
    return 7 / 4 * n + (-1) ** (n + 1) * (5 / 4 * n + 0.5) + 0.5

def get_iterations_to_one(n):
    x = n
    iterations = 0
    while x > 1:
        x = iterate(x)
        iterations += 1
    
    return iterations

def get_iterations_to_power_of_two(n):
    def n_is_power_of_two(n):
        n = int(n)
        
        return (n & (n-1) == 0) and (n != 0 and n-1 != 0)
    
    x = n
    iterations = 0
    while not n_is_power_of_two(x):
        x = iterate(x)
        iterations += 1
    
    return (iterations, x)

iterate(9)
```

```python
import pandas as pd
import matplotlib as plt

n = [i for i in range(1, 150)]
iterations_to_one = [get_iterations_to_one(i) for i in n]
power_of_two_results = [get_iterations_to_power_of_two(i) for i in n]
iterations_to_power_of_two = [result[0] for result in power_of_two_results]
powers_of_two = [result[1] for result in power_of_two_results]

df = pd.DataFrame.from_dict({
    'n': n,
    'to_one': iterations_to_one,
    'to_power_of_two': iterations_to_power_of_two,
    'power_of_two': powers_of_two,
})
df.set_index('n', inplace=True)

df = df[['to_power_of_two']]
df.plot()
df.head()
```
